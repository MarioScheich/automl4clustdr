```python
n_configs = 60
```

```python
# Use all algorithms
from ClusteringCS.ClusteringCS import KMeans
from ClusteringCS.ClusteringCS import KMEANS_WARD_SPACE, KMEANS_SPACE
from ClusteringCS.ClusteringCS import execute_algorithm_from_config
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from statistics import mean
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier as RF
from sklearn.metrics import adjusted_rand_score
from sklearn.metrics import normalized_mutual_info_score
```

```python
c25_stats =[]
mean_measures = ['accuracy', 'f1', 'Adjusted Rand Index', 'normalized_mutual_info_score', 'Runtime in seconds', 'clustering_time']
x = 0
for i in samples:
    data = {"Measure" : ['accuracy', 'f1', 'Adjusted Rand Index', 'normalized_mutual_info_score', 'Runtime in seconds', 'clustering_time', 'predictions', 'clusters', 'configs']}
    c25_stats.append(pd.DataFrame(data))
    means = []
    
    for metric in metrics:
        tic1 = time.perf_counter()
        automl_four_clust_instance = optimizer(dataset=i[0], metric=metric, n_loops=n_configs, cs= KMEANS_WARD_SPACE, optimization= "bayesian", 
                                               data_reduction =  "LaplacianScore", data_reduction_param = {"feature_number": 7, "t_param": 0, "neighbour_size":5})
                                               
        
        
        #"feature_number": 7}) #"t_param": 625, "neighbour_size": 5})

        #"kmeans"
        acc_scores = []
        f1_scores = []
        adjusted_rand_scores = []
        normalized_mutual_info_scores = []
        clusters = []
        configs = []
        automl_times = []
        clustering_times = []
        while len(f1_scores) < 2:
            
            #smac, best_configuration = automl_four_clust_instance.optimize()
            automl_four_clust_instance.optimize()
            #automl_four_clust_instance.get_run_history()
            best_configuration = automl_four_clust_instance.get_incumbent()
            tic2 = time.perf_counter()
            y_best = execute_algorithm_from_config(i[1], best_configuration)
            toc2 = time.perf_counter()
            clustering_time = toc2-tic2
            clustering_times.append(clustering_time)
            best_config = best_configuration.get_dictionary().copy()
            
            try:
                n_clusters = best_config.pop("n_clusters")
            except:
                try:
                    n_clusters = best_config.pop("n_components")
                except:
                    n_clusters = len(np.unique(y_best))+1
            
            used_algorithm = best_config.pop("algorithm")
            configs.append(used_algorithm)
            acc_scores.append(accuracy_score(i[4], y_best))
            f1 = f1_score(i[4], y_best, average="weighted")
            f1_scores.append(f1)
            adjusted_rand_scores.append(adjusted_rand_score(i[4], y_best))
            normalized_mutual_info_scores.append(normalized_mutual_info_score(i[4], y_best))
            clusters.append(n_clusters)
    
        toc1 = time.perf_counter()
        automl_time = toc1-tic1
        automl_times.append(automl_time)
        c25_stats[x][metric.name] = [mean(acc_scores), mean(f1_scores), mean(adjusted_rand_scores), mean(normalized_mutual_info_scores), (sum(automl_times))/2, sum(clustering_times), y_best, clusters, configs]
    
    for i in mean_measures:
        t = c25_stats[x]
        t = t.loc[t['Measure'] == i]
        t = t.drop("Measure", axis=1)
        means.append(t.mean(axis=1).iloc[0])
    means.extend((0, 0, 0))
    c25_stats[x]['mean'] = means
    x = x+1
```
